# **1. What are epoch, batch size, and Stratified? Please explain them.**

### **Epoch:**

An epoch means the neural network has seen the entire training dataset exactly once. During one epoch, the model processes all training samples, computes the loss, and updates its weights using backpropagation. Training is usually done for multiple epochs (e.g., 10, 50, 100) so the model can gradually learn patterns in the data. Fewer epochs may lead to underfitting, while too many epochs can cause overfitting, where the model memorizes the training data instead of generalizing.

### **Batch Size:**

Batch size refers to the number of training samples processed before the model updates its weights.

Instead of updating weights after every single sample (batch size = 1) or after the entire dataset at once (batch size = full dataset), the data is divided into smaller batches.

Small batch size → noisier updates, but faster learning and better generalization

Large batch size → more stable gradients, but requires more memory and may converge to suboptimal solutions

Batch size controls the trade-off between speed, memory usage, and model performance.

### **Stratified:**

Stratified sampling is a technique used when splitting datasets into training, validation, or test sets. It ensures that each split maintains the same class distribution as the original dataset.

This is especially important for imbalanced datasets (e.g., 90% Class A and 10% Class B). Without stratification, one subset might end up with very few or even no samples from the minority class, leading to biased training and unreliable evaluation results.

Stratification ensures fair representation of all classes across all splits, making model evaluation more accurate and reliable.

## **3. How do we handle structured and unstructured data in ANNs and CNNs?**

**Structured Data** (tabular data) refers to data that is organized in a fixed format such as rows and columns, for example CSV files or spreadsheets containing features like age, salary, or credit score. This type of data is best handled using Artificial Neural Networks (ANNs) or Multi-Layer Perceptrons (MLPs), where each feature is treated as an input neuron and the data is converted into a flat numerical vector. The network processes this data through fully connected (dense) layers and learns relationships between features for tasks like classification and regression, without considering any spatial or sequential structure.

**Unstructured Data** (images, text, audio, video) refers to data that does not have a predefined format, such as images, videos, raw text, or sound signals. For image and video data, Convolutional Neural Networks (CNNs) are most suitable because they take input as multi-dimensional tensors (height × width × channels) and use convolution operations to automatically extract important spatial features like edges, textures, and objects in a hierarchical way. This makes CNNs much more efficient than ANNs for visual data, as they preserve spatial relationships and reduce complexity through parameter sharing.

## **4. What is a layer? Define ANN and CNN layers?**

**A layer** in a neural network is a collection of neurons that perform a specific transformation on input data and pass the output to the next layer. Each layer helps the network learn different levels of patterns, starting from simple features in early layers to more complex representations in deeper layers.

An **Artificial Neural Network (ANN)** is a type of neural network designed mainly for structured/tabular data, where information flows through fully connected layers. The main layers in ANN are the input layer (receives feature values), hidden layers (perform weighted computations and apply activation functions to learn patterns), and the output layer (produces final prediction such as class or value). In ANN, every neuron in one layer is connected to every neuron in the next layer.

A **Convolutional Neural Network (CNN)** is designed for unstructured data like images. Its architecture includes specialized layers such as the convolutional layer (extracts features like edges and patterns using filters), the pooling layer (reduces spatial size and computation while retaining important information), and the fully connected layer (performs final classification or prediction). CNN layers preserve spatial relationships in data and automatically learn hierarchical feature representations.

## **5. What concepts are used in ANN and CNN?**

In Artificial Neural Networks (ANNs) and Convolutional Neural Networks (CNNs), several key concepts are used to help the model learn patterns from data effectively.

In ANNs, the main concepts include neurons, weights, and bias, where each input is multiplied by a weight and adjusted using bias to control learning. The activation function is used to introduce non-linearity so the model can learn complex relationships. ANNs also use fully connected layers, where every neuron is connected to every neuron in the next layer. Training is done using forward propagation (to get predictions), loss function (to measure error), and backpropagation with gradient descent (to update weights and reduce error).

In CNNs, similar concepts are used but with additional image-focused operations. CNNs use filters/kernels in the convolution operation to extract features like edges, textures, and shapes. They also use pooling layers (max pooling or average pooling) to reduce dimensionality and computation while keeping important features. Like ANNs, CNNs also use activation functions, loss functions, and backpropagation for learning. Overall, CNNs add spatial feature extraction concepts on top of basic neural network principles.
